# **Solubility Prediction using ML**

# **Load Dataset**

```python
# Importing libraries

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

# **Data Preparation**

## Data Separation as X and y

```python
y = df["logS"]
y
```

```python
X = df.drop("logS", axis = 1)
X
```

```python
# Importing sckit-learn library for splitting the dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
# Training set contains 80% of the data
X_train
```

```python
# Testing set contains 20% of the data
X_test
```

# **Model Building**

## **Linear Regression**

### **Training the model**

```python
# Importing Linear Regression model from scikit-learn library
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)  # Fitting the model to the training set
```

### **Applying the model to make prediction**

```python
# Predicting the training and testing set results
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred  # 80% of the data
```

```python
y_lr_test_pred  # 20% of the data
```

### **Evaluate model performance**

```python
y_train
```

```python
y_lr_train_pred
```

```python
# Importing metrics for evaluating the model performance
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
# Displaying the results
print("LR MSE (Train): ", lr_train_mse)
print("LR R2 (Train): ", lr_train_r2)
print("LR MSE (Test): ", lr_test_mse)
print("LR R2 (Test): ", lr_test_r2)
```

```python
# Creating a DataFrame to summarize the results
lr_results = pd.DataFrame(["Linear Regression", lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ["Method", "Training MSE", "Training R2", "Test MSE", "Test R2"]
```

```python
lr_results
```

## **Random Forest**

### **Training the model**

Here we have quantitative `y`, hence we use Regression model
<!-- if the `y` is quantitative then we build a `regression` model<br>
if the `y` is not quantitative then we build a `classification` model -->

```python
# importing Random Forest Regressor from scikit-learn library
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state = 100)  # Creating an instance of RandomForestRegressor with max_depth of 2
rf.fit(X_train, y_train)
```

### **Applying the model to make prediction**

```python
# Predicting the training and testing set results using Random Forest
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### **Evaluate model performance**

```python
# Importing metrics for evaluating the model performance
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print("RF MSE (Train): ", rf_train_mse)
print("RF R2 (Train): ", rf_train_r2)
print("RF MSE (Test): ", rf_test_mse)
print("RF R2 (Test): ", rf_test_r2)
```

```python
rf_results = pd.DataFrame(["Random Forest", rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ["Method", "Training MSE", "Training R2", "Test MSE", "Test R2"]
```

```python
rf_results
```

## **Model Comparision**

```python
# Concatenating the results of both models into a single DataFrame
df_models = pd.concat([lr_results, rf_results], axis = 0)
df_models
```

```python
df_models.reset_index(drop = True)
```

# **Data Visualization of prediction results**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (5, 5))
plt.scatter(x= y_train, y= y_lr_train_pred, alpha = 0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), "r--")
plt.xlabel("Experimental logS")
plt.ylabel("Predicted logS")
plt.show()
```
